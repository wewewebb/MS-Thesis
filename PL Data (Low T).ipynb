```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.interpolate as spi
import scipy.signal as sps
import matplotlib.patches as pltp

#Low T PL Data (laser)
en_Q46_laser_df = pd.read_excel('New PL Data.xlsx', sheet_name='PL Data (Low T-13 K)', usecols=[5])
Q46_laser_df = pd.read_excel('New PL Data.xlsx', sheet_name='PL Data (Low T-13 K)', usecols=[6])

en_Q40_laser_df = pd.read_excel('New PL Data.xlsx', sheet_name='PL Data (Low T-13 K)', usecols=[1])
Q40_laser_df = pd.read_excel('New PL Data.xlsx', sheet_name='PL Data (Low T-13 K)', usecols=[2])

en_Q44_laser_df = pd.read_excel('New PL Data.xlsx', sheet_name='PL Data (Low T-13 K)', usecols=[9])
Q44_laser_df = pd.read_excel('New PL Data.xlsx', sheet_name='PL Data (Low T-13 K)', usecols=[10])

#To check if you loaded the correct data
#print(en_Q46_laser_df)
#print(Q46_laser_df)
#print(en_Q40_laser_df)
#print(Q40_laser_df)
#print(en_Q44_laser_df)
#rint(Q44_laser_df)

#Low T PL Data (lamp)
en_Q46_lamp_df = pd.read_excel('New PLE Data 2 (bigger focusing lens).xlsx', sheet_name='Q046 PL', usecols=[1])
Q46_lamp_df = pd.read_excel('New PLE Data 2 (bigger focusing lens).xlsx', sheet_name='Q046 PL', usecols=[2])

en_Q40_lamp_df = pd.read_excel('New PLE Data 2 (bigger focusing lens).xlsx', sheet_name='Q040 PL', usecols=[1])
Q40_lamp_df = pd.read_excel('New PLE Data 2 (bigger focusing lens).xlsx', sheet_name='Q040 PL', usecols=[2])

en_Q44_lamp_df = pd.read_excel('New PLE Data 2 (bigger focusing lens).xlsx', sheet_name='Q044 OFF PL', usecols=[1])
Q44_lamp_df = pd.read_excel('New PLE Data 2 (bigger focusing lens).xlsx', sheet_name='Q044 OFF PL', usecols=[2])

#To check if you loaded the correct data
#print(en_Q46_lamp_df)
#print(Q46_lamp_df)
#print(en_Q40_lamp_df)
#print(Q40_lamp_df)
#print(en_Q44_lamp_df)
#rint(Q44_lamp_df)

#Low T PL Data (lamp 2)
en_Q46_lamp_df_2 = pd.read_excel('New PLE Data 3.xlsx', sheet_name='Q046 PL', usecols=[1])
Q46_lamp_df_2 = pd.read_excel('New PLE Data 3.xlsx', sheet_name='Q046 PL', usecols=[2])

en_Q40_lamp_df_2 = pd.read_excel('New PLE Data 3.xlsx', sheet_name='Q040 PL', usecols=[1])
Q40_lamp_df_2 = pd.read_excel('New PLE Data 3.xlsx', sheet_name='Q040 PL', usecols=[2])

en_Q44_lamp_df_2 = pd.read_excel('New PLE Data 3.xlsx', sheet_name='Q044 OFF PL', usecols=[1])
Q44_lamp_df_2 = pd.read_excel('New PLE Data 3.xlsx', sheet_name='Q044 OFF PL', usecols=[2])

#To check if you loaded the correct data
#print(en_Q46_lamp_df_2)
#print(Q46_lamp_df_2)
#print(en_Q40_lamp_df_2)
#print(Q40_lamp_df_2)
#print(en_Q44_lamp_df_2)
#rint(Q44_lamp_df_2)

#Convert dataframe to 1D arrays
en_Q46_laser = en_Q46_laser_df.to_numpy().reshape(-1)
Q46_laser = Q46_laser_df.to_numpy().reshape(-1)
en_Q40_laser = en_Q40_laser_df.to_numpy().reshape(-1)
Q40_laser = Q40_laser_df.to_numpy().reshape(-1)
en_Q44_laser = en_Q44_laser_df.to_numpy().reshape(-1)
Q44_laser = Q44_laser_df.to_numpy().reshape(-1)

#To check if the conversion is correct
#print(en_Q46_laser)
#print(Q46_laser)
#print(en_Q40_laser)
#print(Q40_laser)
#print(en_Q44_laser)
#print(Q44_laser)

en_Q46_lamp = en_Q46_lamp_df.to_numpy().reshape(-1)
Q46_lamp = Q46_lamp_df.to_numpy().reshape(-1)
en_Q40_lamp = en_Q40_lamp_df.to_numpy().reshape(-1)
Q40_lamp = Q40_lamp_df.to_numpy().reshape(-1)
en_Q44_lamp = en_Q44_lamp_df.to_numpy().reshape(-1)
Q44_lamp = Q44_lamp_df.to_numpy().reshape(-1)

#To check if the conversion is correct
#print(en_Q46_lamp)
#print(Q46_lamp)
#print(en_Q40_lamp)
#print(Q40_lamp)
#print(en_Q44_lamp)
#print(Q44_lamp)

en_Q46_lamp_2 = en_Q46_lamp_df_2.to_numpy().reshape(-1)
Q46_lamp_2 = Q46_lamp_df_2.to_numpy().reshape(-1)
en_Q40_lamp_2 = en_Q40_lamp_df_2.to_numpy().reshape(-1)
Q40_lamp_2 = Q40_lamp_df_2.to_numpy().reshape(-1)
en_Q44_lamp_2 = en_Q44_lamp_df_2.to_numpy().reshape(-1)
Q44_lamp_2 = Q44_lamp_df_2.to_numpy().reshape(-1)

#To check if the conversion is correct
#print(en_Q46_lamp_2)
#print(Q46_lamp_2)
#print(en_Q40_lamp_2)
#print(Q40_lamp_2)
#print(en_Q44_lamp_2)
#print(Q44_lamp_2)

#Making a spline of low T PL Data (laser)
en_Q46_laser_new = np.linspace(en_Q46_laser.min(), en_Q46_laser.max(), 3000)
spl1 = spi.make_interp_spline(en_Q46_laser, Q46_laser, k=3)
Q46_laser_new = spl1(en_Q46_laser_new)

en_Q40_laser_new = np.linspace(en_Q40_laser.min(), en_Q40_laser.max(), 3000)
spl2 = spi.make_interp_spline(en_Q40_laser, Q40_laser, k=3)
Q40_laser_new = spl2(en_Q40_laser_new)

en_Q44_laser_new = np.linspace(en_Q44_laser.min(), en_Q44_laser.max(), 3000)
spl3 = spi.make_interp_spline(en_Q44_laser, Q44_laser, k=3)
Q44_laser_new = spl3(en_Q44_laser_new)

#Making a spline of low T PL Data (lamp)
en_Q46_lamp_new = np.linspace(en_Q46_lamp.min(), en_Q46_lamp.max(), 3000)
spl4 = spi.make_interp_spline(en_Q46_lamp, Q46_lamp, k=3)
Q46_lamp_new = spl4(en_Q46_lamp_new)

en_Q40_lamp_new = np.linspace(en_Q40_lamp.min(), en_Q40_lamp.max(), 3000)
spl5 = spi.make_interp_spline(en_Q40_lamp, Q40_lamp, k=3)
Q40_lamp_new = spl5(en_Q40_lamp_new)

en_Q44_lamp_new = np.linspace(en_Q44_lamp.min(), en_Q44_lamp.max(), 3000)
spl6 = spi.make_interp_spline(en_Q44_lamp, Q44_lamp, k=3)
Q44_lamp_new = spl6(en_Q44_lamp_new)

#Making a spline of low T PL Data (lamp 2)
en_Q46_lamp_new_2 = np.linspace(en_Q46_lamp_2.min(), en_Q46_lamp_2.max(), 3000)
spl4_2 = spi.make_interp_spline(en_Q46_lamp_2, Q46_lamp_2, k=3)
Q46_lamp_new_2 = spl4_2(en_Q46_lamp_new_2)

en_Q40_lamp_new_2 = np.linspace(en_Q40_lamp_2.min(), en_Q40_lamp_2.max(), 3000)
spl5_2 = spi.make_interp_spline(en_Q40_lamp_2, Q40_lamp_2, k=3)
Q40_lamp_new_2 = spl5_2(en_Q40_lamp_new_2)

en_Q44_lamp_new_2 = np.linspace(en_Q44_lamp_2.min(), en_Q44_lamp_2.max(), 3000)
spl6_2 = spi.make_interp_spline(en_Q44_lamp_2, Q44_lamp_2, k=3)
Q44_lamp_new_2 = spl6_2(en_Q44_lamp_new_2)

#Font sizes for the plot texts
SMALL_SIZE = 12
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

#Setting font sizes for plot elements
plt.rc('font', size=SMALL_SIZE) #controls default text sizes
plt.rc('axes', labelsize=BIGGER_SIZE) #font size of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE) #font size of the x tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE) #font size of the y tick labels
plt.rc('legend', fontsize=SMALL_SIZE) #legend font size

#Spline plot of Q046 data
plt.plot(en_Q46_laser_new, Q46_laser_new, color='xkcd:magenta', label='laser')
plt.plot(en_Q46_lamp_new, Q46_lamp_new, color='xkcd:blue green', label='lamp (1st run)')
plt.plot(en_Q46_lamp_new_2, Q46_lamp_new_2, color='xkcd:indigo', label='lamp (2nd run)')

#Finding peaks of Q046 spline data
peaksQ46_laser, _ = sps.find_peaks(Q46_laser_new, height=10, prominence=0.01)
#print(en_Q46_laser_new[peaksQ46_laser])
#print(Q46_laser_new[peaksQ46_laser])

peaksQ46_lamp, _ = sps.find_peaks(Q46_lamp_new, height=10, prominence=0.01)
#print(en_Q46_lamp_new[peaksQ46_lamp])
#print(Q46_lamp_new[peaksQ46_lamp])

peaksQ46_lamp_2, _ = sps.find_peaks(Q46_lamp_new_2, height=10, prominence=0.01)
#print(en_Q46_lamp_new_2[peaksQ46_lamp_2])
#print(Q46_lamp_new_2[peaksQ46_lamp_2])

#Plot of Q046 data peaks
plt.plot(en_Q46_laser_new[peaksQ46_laser], Q46_laser_new[peaksQ46_laser], "o", color='xkcd:sea blue')
plt.plot(en_Q46_lamp_new[peaksQ46_lamp], Q46_lamp_new[peaksQ46_lamp], "o", color='xkcd:crimson')
plt.plot(en_Q46_lamp_new_2[peaksQ46_lamp_2], Q46_lamp_new_2[peaksQ46_lamp_2], "o", color='xkcd:tomato')

#Labelling peaks with values
for i, txt in enumerate(np.round(en_Q46_laser_new[peaksQ46_laser],3)):
    plt.annotate(txt, (en_Q46_laser_new[peaksQ46_laser][i], Q46_laser_new[peaksQ46_laser][i]), 
                 xytext=(en_Q46_laser_new[peaksQ46_laser][i]+0.01, Q46_laser_new[peaksQ46_laser][i]-5), 
                 rotation=0, arrowprops=dict(facecolor='black', headwidth=6, width=1, shrink=0.05))

for i, txt in enumerate(np.round(en_Q46_lamp_new[peaksQ46_lamp],3)):
    plt.annotate(txt, (en_Q46_lamp_new[peaksQ46_lamp][i], Q46_lamp_new[peaksQ46_lamp][i]), 
                 xytext=(en_Q46_lamp_new[peaksQ46_lamp][i]+0.01, Q46_lamp_new[peaksQ46_lamp][i]-5), 
                 rotation=0, arrowprops=dict(facecolor='black', headwidth=6, width=1, shrink=0.05))

for i, txt in enumerate(np.round(en_Q46_lamp_new_2[peaksQ46_lamp_2],3)):
    plt.annotate(txt, (en_Q46_lamp_new_2[peaksQ46_lamp_2][i], Q46_lamp_new_2[peaksQ46_lamp_2][i]), 
                 xytext=(en_Q46_lamp_new_2[peaksQ46_lamp_2][i]+0.01, Q46_lamp_new_2[peaksQ46_lamp_2][i]-5), 
                 rotation=0, arrowprops=dict(facecolor='black', headwidth=6, width=1, shrink=0.05))

#Making patches for legend
patch_1 = pltp.Patch(label='laser (16μm)', color='xkcd:magenta')
patch_2 = pltp.Patch(label='lamp (1st run, 30μm)', color='xkcd:blue green')
patch_3 = pltp.Patch(label='lamp (2nd run, 30μm)', color='xkcd:indigo')

plt.gcf().set_size_inches(7, 6) #sets the size of the plot
plt.legend(handles=[patch_1, patch_2, patch_3], loc='upper left') #inserts legend in the plot
plt.xlabel('Energy (eV)') #label for the x-axis
plt.ylabel('PL Intensity (arb. units)') #label for the y-axis
plt.xlim(en_Q46_laser.min(), en_Q46_laser.max()) #sets limit values for x-axis
plt.ylim(-5,300) #sets limit values for y-axis
plt.xticks(np.arange(1.48,1.58,step=0.02)) #sets x-axis tick values
plt.yticks(np.arange(0,301,step=50)) #sets y-axis tick values
#plt.savefig('Q046_PL_comparison.png', dpi=2000, bbox_inches='tight') #saves the figure as png
plt.show()
```
